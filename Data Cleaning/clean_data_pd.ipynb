```python
#Cleaning the dataset using Pandas
import pandas as pd
#Load the datasets1
df1 = pd.read_csv('data.csv')

df1
```

```python
#Load the Datasets2
df2 = pd.read_csv("data1.csv")

df2
```

```python
#concatenate two different dataset
new_df = pd.concat([df1,df2], ignore_index=True)

new_df
```

```python
#Remove all rows with NULL values
new_df.dropna()
```

```python
#index is re-assigned starting from 0 after rows are dropped
new_df.dropna(ignore_index=True)
```

```python
#Replace NULL values with the number 130
new_df.fillna(100)
```

```python
#Replace NULL values in the "Pulse" columns with the number 100
new_df['Pulse'].fillna(100)
```

```python
#Convert Date into correct format 
new_df['Date'] = pd.to_datetime(new_df['Date'])

new_df
```

```python
"""
Loop through all values in the "Duration" column.

If the value is higher than 120, set it to 120:
"""

for i in new_df.index:
    if new_df.loc[i,"Duration"] >= 120:
        new_df.loc[i,"Duration"] = 120

new_df
```

```python
#Print True if Duplicate is Found, Otherwise False
new_df.duplicated()
```

```python
#Remove the Duplicate Values and from Original DataFrame 
new_df.drop_duplicates(inplace=True,ignore_index=True)

new_df
```
